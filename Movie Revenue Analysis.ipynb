### Drop columns without neccesary information and remove all records with no financial information -- Pay close attention to things that don't tell you anything regarding financial data

```python
#needed = [df['id'], df['budget'], df['revenue'], df['original_title'], df['director'], df['genres'], df['production_companies'], df['budget_adj'], df['revenue_adj']]
id = df['id']
budget = df['budget']
revenue = df['revenue']
original_title = df['original_title']
director = df['director']
genres = df['genres']
production_companies = df['production_companies']
budget_adj = df['budget_adj']
revenue_adj = df['revenue_adj']

imdb_go = {'id': id, 'budget': budget, 'revenue': revenue, 'original_title': original_title, 'director': director, 'genres': genres, 'production_companies': production_companies, 'budget_adj': budget_adj, 'revenue_adj': revenue_adj}
imdb = pd.DataFrame.from_dict(imdb_go)
imdb.dtypes
# imdb.groupby('director', as_index = False).head()
```

```python
#joels' version of data cleaning
df.drop(df.query('budget==0 or revenue == 0').index , inplace=True)
df.drop(['id', 'popularity', 'imdb_id', 'homepage', 'budget', 'runtime', 'tagline', 'keywords', 'vote_count', 'vote_average', 'release_date', 'budget_adj', 'overview'], axis=1, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
#separating out production companies on the pipe
production_records = df[df.production_companies.str.contains('|')]
production_records.production_companies = production_records.production_companies.apply(lambda x: x.split("|"))
production_records

df_pc = pd.DataFrame(columns = df.columns)
    for i in range(len(production_records)):
        record = production_records.iloc[i] #returns all data for every single record with abiility to access columns
        for production_company in production_records.production_companies[i]: #loops through all prod compannies in each record
            df_pc = df_pc.append(pd.DataFrame([record.revenue, record.original_title, record.cast, record.director, record.genres, production_company, record.release_year, record.revenue_adj], columns = df.columns))

df_pc.drop('genres', axis=1, inplace=True)
df_pc.reset_index(drop=True, inplace=True)

df_pc.head()
```

## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
prods = imdb.groupby(['production_companies', 'id'])
prods.count().tail()
```

### What 5 movie genres grossed the highest all-time?

```python
gross = imdb.groupby(['genres', 'revenue_adj'])
gross.sum().sort_values('revenue_adj').tail()
```

### Who are the top 5 grossing directors?

```python
directs = imdb.groupby(['revenue_adj', 'director'])
directs.sum().tail(6)
#looks like james cameron, george lucas, james cameron, william
#friedkin, steven spielberg, jj abrams
#joel says this is right
```

### Compare the revenue of the highest grossing movies of all time.

```python
high = imdb.groupby(['revenue_adj', 'original_title'])
high.count().tail()
#uhh i think it's avatar, star wars, titanic, exorcist, jaws
```
